#**1 - iris dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the iris dataset
iris_data = pd.read_csv("/content/iris.data", header=None)
iris_data.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
```

```python
# 2D scatter plots of the four features (creates a matrix of scatter plots for each pair of the features)
sns.pairplot(iris_data, hue="class")
plt.show()
```

```python
iris_data["class"] = pd.Categorical(iris_data["class"]).codes
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris_data["sepal_length"], iris_data["sepal_width"], iris_data["petal_width"], c=iris_data["class"])
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
ax.set_zlabel("petal_width")
plt.show()
```

```python
# Visualization of the feature matrix (column 1-4) as an image
sns.heatmap(iris_data.iloc[:, :-1].corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
# For each class, generate histograms for the four features
sns.set(style="darkgrid")
for feature in ["sepal_length", "sepal_width", "petal_length", "petal_width"]:
    g = sns.FacetGrid(iris_data, col="class")
    g.map(plt.hist, feature, bins=10)
    plt.show()
```

```python
# For each class, generate boxplots of the four features
sns.set(style="ticks")
for feature in ["sepal_length", "sepal_width", "petal_length", "petal_width"]:
    sns.boxplot(x="class", y=feature, data=iris_data)
    plt.show()
```

```python
# For each class, calculate the correlation matrix of the four features
correlation_matrices = iris_data.groupby("class").corr()
print("Correlation matrices for each class:")
print(correlation_matrices)
```

```python
# For each class, calculate the covariance matrix of the four features
covariance_matrices = iris_data.groupby("class").cov()
print("\nCovariance matrices for each class:")
print(covariance_matrices)
```

```python
# Create a parallel coordinate plot of the four features for each class and visualize in the same figure
plt.figure(figsize=(12, 6))
pd.plotting.parallel_coordinates(iris_data, "class")
plt.show()
```

2 - Feature matrix normalization


3 - For a new iris data sample with a feature vector of
**S = [4.50, 3.50, 1.46, 0.24]. **

```python
import numpy as np
import pandas as pd
from scipy.spatial import distance
import matplotlib.pyplot as plt
```

```python
# Separate features and labels
X = iris_data.iloc[:, :-1].values

# Normalize the feature matrix
X_normalized = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
# Calculate correlation matrices for each class
correlation_matrices = {}
for cls in iris_data['class'].unique():
    cls_data = iris_data[iris_data['class'] == cls].iloc[:, :-1].values
    cls_normalized = (cls_data - cls_data.mean(axis=0)) / cls_data.std(axis=0)
    correlation_matrices[cls] = np.corrcoef(cls_normalized.T)
```

```python
# Compare correlation matrices before and after normalization
before_normalization = np.corrcoef(X.T)
after_normalization = np.mean([correlation_matrices[cls] for cls in iris_data['class'].unique()], axis=0)
is_same = np.allclose(before_normalization, after_normalization)

```

```python
# New iris data sample
S = np.array([4.50, 3.50, 1.46, 0.24])

# Calculate Minkowski distances with r=1
minkowski_distances = [distance.minkowski(sample, S, p=1) for sample in X_normalized]
```

```python
# Plot Minkowski distances
plt.figure(figsize=(10, 5))
plt.plot(minkowski_distances)
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Minkowski Distances with r=1')
plt.show()
```

```python
# Find the index of the nearest neighbor
nearest_neighbor_index = np.argmin(minkowski_distances)
nearest_neighbor_class = iris_data.loc[nearest_neighbor_index, 'class']

print(f"Nearest neighbor index: {nearest_neighbor_index}")
print(f"Nearest neighbor class: {nearest_neighbor_class}")
```

# **4 - (PCA) on iris dataset**




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python

# Separate features and labels
X = iris_data.iloc[:, :-1].values
y = iris_data.iloc[:, -1].values

# Standardize the features
X_standardized = StandardScaler().fit_transform(X)
```

```python
# Perform PCA
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_standardized)
```

```python
X = iris_data.drop("class", axis=1)
y = iris_data["class"]
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X)
```

```python
# Plotting the 2D scatter plot matrix
plt.figure(figsize=(12, 8))
pd.plotting.scatter_matrix(iris_data, c=y, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
plt.suptitle('2D Scatter Plot Matrix of PCA Components')
plt.show()
```

```python
# 2) Create a 3D scatter plot of the first three components
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
for cls in iris_data['class'].unique():
    ax.scatter(X_pca[y==cls, 0], X_pca[y==cls, 1], X_pca[y==cls, 2], label=cls)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.legend()
plt.show()
```

```python
# 3) Compute the ratio between the variance of each component and the total variance
variance_ratios = pca.explained_variance_ratio_
variance_percentage = variance_ratios * 100

# Visualize the variance percentage of each component in a bar plot
plt.bar(range(1, 5), variance_percentage, tick_label=[f'PC{i}' for i in range(1, 5)])
plt.xlabel('Principal Component')
plt.ylabel('Variance Percentage')
plt.title('Variance Percentage of Each Principal Component')
plt.show()
```

```python
# 4) Calculate the correlation matrix of the four components
correlation_matrix = np.corrcoef(X_pca.T)
```

```python
# 5) Calculate the covariance matrix of the four components
covariance_matrix = np.cov(X_pca.T)

print("Correlation Matrix:")
print(correlation_matrix)
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from sklearn.feature_selection import mutual_info_regression

# Load the dataset
df = pd.read_csv('HW2_data.txt')

# Display the head of the dataset
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Create a matrix of scatter plots
fig = plt.figure(figsize=(10, 8))
sns.pairplot(df)
plt.title('Pairplot of the Four Features')
plt.show()
```

```python
print(df.dtypes)
```

```python
# Convert object columns to numeric
df = df.apply(pd.to_numeric, errors='coerce')

def calculate_t_statistics(df):
    t_statistics = {}
    for i in range(len(df.columns)):
        for j in range(i+1, len(df.columns)):
            feature1 = df.columns[i]
            feature2 = df.columns[j]
            t_stat, _ = ttest_ind(df[feature1], df[feature2])
            t_statistics[(feature1, feature2)] = t_stat
    return t_statistics

t_statistics = calculate_t_statistics(df)
print(t_statistics)
```
